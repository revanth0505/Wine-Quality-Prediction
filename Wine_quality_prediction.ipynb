```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
data=pd.read_csv('/content/WineQT.csv')
data
```

```python
data=data.replace('?',0)
data
```

```python
data.notnull()
```

```python
x = data.drop('quality',axis=1)
y = data['quality'].apply(lambda y_value: 1 if y_value>=7 else 0)
y
```

```python
from sklearn.preprocessing import MinMaxScaler
sclar=MinMaxScaler()
x=sclar.fit_transform(x)
x
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.3)
```

```python
X_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
import time
neighbour = KNeighborsClassifier(n_neighbors=3)
start=time.time()
neighbour.fit(x,y)
score=neighbour.score(X_train,y_train)
print('KNN Score:',score)
stop=time.time()
time_taken=stop-start
print('Time for Training:',time_taken)
```

```python
pred=neighbour.predict(X_test)
accuracy=accuracy_score(y_test,pred)
print("Accuracy score : ",accuracy)
```

```python
li=['KNN Classifier',score,time_taken,accuracy]
li
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,pred))
```
